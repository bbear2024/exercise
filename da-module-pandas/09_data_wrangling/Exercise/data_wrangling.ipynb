### Imports

```python
# import pandas
import pandas as pd
```

The original data directly from gapminder comes in wide format.

Read the `gapminder_total_fertility.csv` file in a dataframe

```python
df = pd.read_csv('../data/gapminder_total_fertility.csv')
```

```python
df
```

In the original gapminder dataset the columns are the country names and years.



![Screenshot%202024-04-15%20at%2013.09.45.png](./Screenshot%202024-04-15%20at%2013.09.45.png)

Melt the table into a form that resembles the data below:



![Screenshot%202024-04-15%20at%2013.10.23.png](./Screenshot%202024-04-15%20at%2013.10.23.png)

```python
df_long = df.melt(id_vars='country',
                  var_name='year',
                  value_name='fertility_rate'
                  )
```

```python
df_long
```

Pivot the table back to its original form.

```python
df_wide = pd.pivot(df_long,
                   index='country',
                   columns='year',
                   values='fertility_rate')
```

```python
df_wide.reset_index()
```

Create an aggreate table with the average fertility_rate of each country.



![Screenshot%202024-04-15%20at%2013.18.30.png](./Screenshot%202024-04-15%20at%2013.18.30.png)

```python
df_wide_mean = df_long.pivot_table(columns='country',
                                   values='fertility_rate',
                                   aggfunc='mean')
df_wide_mean
```
