# Visualize the names dataset
Create a plot of the summed up frequency (total births) per year from 1880 to 2020.

Step 1 - Create the data

Make a empty dictionary that maps the column names year and total_births to empty lists:

births_dict = {'year':[], 'total_births':[]}

Loop over the years from 1880 to 2020

Use your parse_dataset(year) function to parse each year

Use the pandas.sum() method to calculate the sum of the frequencies (births)

Append the result to the dictionary using births_dict['total_births'].append()

Also append the year to births_dict['year']

Create a yearly_births DataFrame by passing births_dictto pd.DataFrame()

This should result in a DataFrame that has the year as one column and the total_birthsin that year in another column.



Step 2 - Plotting

use pandas to make a quick plot of your results

use matplotlib to make a more detailed plot but adding a legend, title and and axis labels

```python
def parse_dataset(year):
    import pandas as pd
    df = pd.read_csv(f'../data/names/yob{year}.txt', names=['name','gender','frequency'], header=None)
    return df
```

```python
births_dict = {'year':[], 'total_births':[]}
```

```python
for year in range(1880,2023):
    df = parse_dataset(year)
    births_dict['year'].append(year)
    births_dict['total_births'].append(df.frequency.sum())
```

```python
import pandas as pd
```

```python
yearly_births = pd.DataFrame(births_dict)
```

```python
yearly_births.head()
```

```python
yearly_births.plot(kind='area', x='year');
```

```python
import matplotlib.pyplot as plt
x=yearly_births['year']
y=yearly_births['total_births']
plt.bar(x,y,)
plt.xlabel('Year')
plt.ylabel('Total Births')
plt.title('Yearly Births')
```

# Visualize the Distribution of Names

Read in a single dataset for one year using the parse_dataset() function and create a histogram of the frequency column.

The distribution is highly skewed! Include the following log=True in plt.hist to improve the visualization!

Play around with the number of bins bins to make the visualization even clearer.

Do the same for another year and compare the two distributions with each other.

Add labels and a title

```python
from matplotlib import pyplot as plt
```

```python
df_2019 = parse_dataset(2019)
plt.hist(df_2019['frequency']);
```

```python
plt.hist(df_2019['frequency'], log=True);
```

```python
plt.hist(df_2019['frequency'], log=True, bins=100)
plt.title('Distribution of Names in Year 2019')
plt.xlabel('Frequency')
plt.ylabel('Number of Names')
```

```python
df_1904 = parse_dataset(1904)
plt.hist(df_1904['frequency'], log=True, bins=100)
plt.xlabel('Frequency')
plt.ylabel('Number of Names')
plt.title('Distribution of Names in Year 1904')
```

```python
male_2019 = df_2019[df_2019['gender'] == 'M']
female_2019 = df_2019[df_2019['gender'] == 'F']
listof2019 = [male_2019['frequency'], female_2019['frequency']]
plt.hist(x = listof2019, label = ['Male','Female'], log=True)
plt.legend()
plt.xlabel('Frequency')
plt.ylabel('Number of Names')
plt.title('Distribution of Names in Year 2019')
```
