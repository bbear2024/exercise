```python
import pandas as pd
```

```python
def clean_bike_data(df):
    
        # drop duplicates
    df = df.drop_duplicates()
        # drop column 'angelegt_am' and 'erfassungsgrund' - irrelevant to us, when and why observation got added to the database.
    df = df.drop('angelegt_am', axis='columns')
    df = df.drop('erfassungsgrund', axis='columns')
        # we have just 167 attempts and 7 thefts of unknown state in our dataset, so we decide to drop those observations.
    df = df[df['versuch'].isin(['Nein'])]
        # change date text string to datetime datatype

    df['tatzeit_anfang_datum'] = pd.to_datetime(df['tatzeit_anfang_datum'],format='%d.%m.%Y')
    df['tatzeit_ende_datum'] = pd.to_datetime(df['tatzeit_ende_datum'],format='%d.%m.%Y')
    return df
```

```python
df2020 = pd.read_csv('bike_theft_berlin_2020.csv', encoding='latin-1') 
 # make column names lowercase
df2020.columns = df2020.columns.str.lower() 

df2020
```

```python
cdf2020 = clean_bike_data(df2020)
cdf2020
```

```python
df2021 = pd.read_csv('bike_theft_berlin_2021.csv', encoding='latin-1') 
 # make column names lowercase
df2021.columns = df2021.columns.str.lower() 

cdf2021 = clean_bike_data(df2021)
cdf2021
```

```python
df2022 = pd.read_csv('bike_theft_berlin_2022.csv', encoding='latin-1') 
 # make column names lowercase
df2022.columns = df2022.columns.str.lower() 

cdf2022 = clean_bike_data(df2022)
cdf2022
```

```python
df2023 = pd.read_csv('bike_theft_berlin_2023.csv', encoding='latin-1') 
 # make column names lowercase
df2023.columns = df2023.columns.str.lower() 

cdf2023 = clean_bike_data(df2023)
cdf2023
```

```python
completelist = pd.concat([cdf2020,cdf2021,cdf2022,cdf2023], ignore_index=True)
completelist.duplicated(keep=False).value_counts() # check if have duplicates
```

```python
completelist.drop_duplicates(keep='first', inplace=True, ignore_index= True) #drop duplicates
```

```python
completelist.sort_values(by=['tatzeit_anfang_datum','tatzeit_ende_datum','tatzeit_ende_stunde'], inplace=True) #sort by date and time that crime committed
completelist
```

```python
original_lor = pd.read_excel('broker.xls') 
original_lor
```

```python
small_lor = original_lor[['Planungsraum (Nummer)','Bezirk','Bezirksregion (Name)','Planungsraum (Name)']]
small_lor.rename(columns={'Planungsraum (Nummer)':'lor'}, inplace=True)
small_lor.columns = small_lor.columns.str.lower()

small_lor
```

```python
finish_list = pd.merge(left= completelist,right= small_lor,how='left',on='lor')
finish_list
```

```python
# Translate column into English
finish_list.columns = ['crime_time_start_date','crime_time_start_hour','crime_time_end_date','crime_time_end_hour','lor','amount_of_damage','attempt','type_of_bicycle','offense/crime','district','district_region','planning_area']
```

```python
# drop 'attempt' column because all values are same
finish_list.drop('attempt', axis=1, inplace=True)
```
