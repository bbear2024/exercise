4. Add the sum of the sepal width and length as a new column to your data frame.

```python
iris['sum_sepal_width_length'] = iris['sepal_width'] + iris['sepal_length']
iris
```

5. Create a new column with a rough estimate of petal area by multiplying petal length and width together.

```python
iris['estimate_petal_area'] = iris['petal_length'] * iris['petal_width']
iris
```

6. Create a new dataframe with petal areas greater than $1cm^2$.

```python
iris_big = iris[iris['estimate_petal_area'] > 1]
iris_big
```

7. Using the original unfiltered dataframe, create 3 new dataframes, each containing only irises of each a single species 'Iris-setosa', 'Iris-versicolor' or 'Iris-virginica'.

```python
iris_setosa = iris[iris['species'] == 'setosa']
iris_versicolor = iris[iris['species'] == 'versicolor']
iris_virginica= iris[iris['species'] == 'virginica']
```

5. Now select all the rows where the `chlorides` value is greater than the column's mean

```python
wine[(wine['pH'] > 3.0) & (wine['pH'] < 3.5)]
```

7. Further filter the results from 6 to grab only those rows that have a `residual sugar` less than 2.0.

```python
wine[(wine['pH'] > 3.0) & (wine['pH'] < 3.5) & (wine['residual sugar'] < 2.0)]
```
